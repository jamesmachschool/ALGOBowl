```python
import random

class InputGenerator:
    
    def __init__(self, n, m):
        self.n = n
        self.m = m
        self.data = ""
        
    def generate_new_input(self):
        self.data = self.data + str(self.n) + "\n"
        self.data = self.data + str(self.m) + "\n"
        usednumbers = set()
        
        for i in range(self.m-1):
            randsubset = self.generate_random_subset()
            randsubset.sort()
            for j in randsubset:
                usednumbers.add(j)
                self.data = self.data + str(j) +" "
            self.data = self.data +"\n"+str(random.randint(1,self.n+1))+"\n"
        
        theRest = self.subset_not_including(usednumbers)
        for i in theRest:
            self.data = self.data + str(i) +" "
        self.data = self.data +"\n"+str(random.randint(1,self.n+1))+"\n"
        
    def generate_random_subset(self, size = None):
        
        if size == None:
            size = random.randint(1,self.n+1)
        
        subset = []
        for i in range(size):
            subset.append(random.randint(1,self.n+1))
            
        return subset
    
    def subset_not_including(self, thisset):
        subset = []
        for i in range(1,self.n):
            if  not i in thisset:
                subset.append(i)
        return subset
        
    def get_data(self):
        return str(self.data)
    
    def write_to_file(self, filename):
        file = open(filename, "w")
        file.write(self.data)
        file.flush()
        file.close()
    
```

```python
#test
ig = InputGenerator(100,78)
ig.generate_new_input()
print(ig.get_data())
ig.write_to_file("test_new_input.txt")
```
